```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

historial = pd.read_csv('/content/sample_data/all-states-history.csv')
datos = pd.DataFrame(historial)

#Total deaths confirmed by state (top 10)
print("Total deaths confirmed by state \n")
print(datos.groupby(["state"])["deathConfirmed"].sum().sort_values(ascending=False).head(10))
print("======================================== \n")

#Highest number of hospitalized by state (top 10)
print("Highest number of hospitalized by state \n")
print(datos.groupby(["state"])["hospitalized"].max().sort_values(ascending=False).head(10))
print("======================================== \n")

#Detailed description of the data in the "deathConfirmed" column
print("Detailed description of the data in the deathConfirmed column \n")
print(datos["deathConfirmed"].describe())
print("======================================== \n")


#Pie chart
print("Pie chart of the total confirmed deaths (top 5)")
df = datos.groupby(["state"])["deathConfirmed"].sum().sort_values(ascending=False)
df.head(5).plot.pie(subplots=True,autopct="%0.1f %%")
plt.show()
print("\n")

#Graph bar
print("Graph bar of the maximum hospitalized (top 10)")
df = datos.groupby(["state"])["hospitalized"].max().sort_values(ascending=False)
df.head(10).plot.barh()
plt.show()
```
